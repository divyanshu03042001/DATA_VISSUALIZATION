```python
#1. Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

#
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Generate x and y values
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Output to notebook
output_notebook()

# Create a new plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='y')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="sin(x)", line_width=2)

# Show the results
show(p)
```

```python
#2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

#
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Generate random data
np.random.seed(42)
x = np.random.rand(100) * 10
y = np.random.rand(100) * 10
sizes = np.random.rand(100) * 20 + 10  # Sizes between 10 and 30
colors = np.random.choice(['red', 'blue', 'green', 'orange', 'purple'], 100)

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Output to notebook
output_notebook()

# Create a new plot
p = figure(title="Random Scatter Plot", x_axis_label='x', y_axis_label='y')

# Add a scatter renderer with sizes and colors
p.scatter('x', 'y', size='sizes', color='colors', source=source, legend_field='colors', fill_alpha=0.6)

# Show the results
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Output to notebook
output_notebook()

# Create a new plot with a title and axis labels
p = figure(x_range=fruits, title="Fruit Counts",
           toolbar_location=None, tools="")

# Add a bar renderer with the data
p.vbar(x=fruits, top=counts, width=0.9)

# Customize the plot
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.xaxis.axis_label = "Fruits"
p.yaxis.axis_label = "Counts"

# Show the results
show(p)
```

```python
#
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Generate data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(hist=hist, left=edges[:-1], right=edges[1:]))

# Output to notebook
output_notebook()

# Create a new plot
p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency')

# Add a quad renderer with the histogram data
p.quad(bottom=0, top='hist', left='left', right='right', source=source, fill_color='blue', line_color='white')

# Show the results
show(p)
```

```python
#
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show, output_file
from bokeh.models import ColorBar, LinearColorMapper
from bokeh.transform import transform
from bokeh.io import output_notebook
from bokeh.layouts import layout

# Provided dataset
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Convert the data to a DataFrame
df = pd.DataFrame(data_heatmap, columns=x, index=y)

# Prepare the data for Bokeh
df = df.stack().reset_index()
df.columns = ['y', 'x', 'value']

# Create a color mapper
mapper = LinearColorMapper(palette="Viridis256", low=df.value.min(), high=df.value.max())

# Create the figure
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', tools="hover", tooltips=[('Value', '@value')])

# Add the heatmap
p.rect(x='x', y='y', width=1/len(x), height=1/len(y), source=df, fill_color=transform('value', mapper), line_color=None)

# Add the color bar
color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
output_notebook()
show(p)
```
